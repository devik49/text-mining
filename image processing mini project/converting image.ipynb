```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image in BGR format
path = r'D:\SportsPersonClassifier\model\test_images\sharapova1.jpg'
img = cv2.imread(path)

```

```python
# Convert the image to RGB format
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

```

```python
# Invert the color values of the image
inverted_img = np.invert(img_rgb)

```

```python
# Display the original and inverted images
plt.subplot(121), plt.imshow(img_rgb)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(inverted_img)
plt.title('Inverted Image'), plt.xticks([]), plt.yticks([])
plt.show()

```

```python
# Save the inverted image as a new file
cv2.imwrite('inverted_image.png', cv2.cvtColor(inverted_img, cv2.COLOR_RGB2BGR))
```
